```python
import csv
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python

# Load CSV file
df = pd.read_csv("transition_data_SI.csv")
df.sample(5)

# Replace multiple column names using a dictionary
```

```python
df.rename(columns={
    "Altitude_km": "Altitude",
    "V??(km/sec)": "Vin",
    "M??": "Mach",
    "rn(mm)": "Nose Radius",
    "? (deg)": "Degree",
    "?c , eff (deg)":"Effective Cone Angle",
    "st(m)":"Transition Start Location",
    "Me":"Edge Mach Number",
    "Local Re (Rst)":"Local Re",
    "Local unit Re (Ri) per m":"Local Re upl",
    "st (m) (end)":"Loc TE",
    "Me (end)":"Mach No at Location",
    "Local Re (end)":"Re at TE"
    
}, inplace=True)

# Save back to CSV (overwrite or new file)
df.to_csv("data_renamed.csv", index=False)

print("Column names updated successfully!")
```
